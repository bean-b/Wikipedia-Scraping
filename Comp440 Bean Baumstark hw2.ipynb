```python
def getArticleText(lang, titles):
    """
    Given a langCode ("en", "de", etc.) and a list of article titles
    Returns a list of strings containing the text of the specified
    articles in the specified language.
  """
    result = []
    for t in titles:
        params = {
            'action' : 'query',
            'prop' : 'extracts',
            'format' : 'json',
            'explaintext' : '',
            'titles' : t
        }
        r = requests.get("https://" + lang + ".wikipedia.org/w/api.php", params=params)
        js = r.json()
        page = next(iter(js['query']['pages'].values()))
        if 'extract' not in page or not page['extract']:
          raise KeyError('No article found with title ' + repr(t) + ' check url ' + r.url)
        result.append(page['extract'])
    return result

articleTexts = (getArticleText("en", wikiRadiationList))
```

```python
r = requests.get('https://raw.githubusercontent.com/stanfordnlp/CoreNLP/master/data/edu/stanford/nlp/patterns/surface/stopwords.txt')
stopwords = set(r.text.split())
print(list(stopwords))
stopwords.add('===')
stopwords.add('==')
stopwords.add('\\displaystyle')
stopwords.add('{\\displaystyle')
```

```python
def make_tf(text):
    counts = Counter()
    for token in text.lower().split():
        if(token not in stopwords):
          counts[token] += 1
    total = sum(counts.values())
    for token in counts:
        counts[token] /= total
    return counts

tf_matrix = np.array([ make_tf(text) for text in articleTexts ])
```

```python
import math

num_docs_per_word = Counter()
for tf_vector in tf_matrix:
    for word in tf_vector:
        num_docs_per_word[word] += 1

num_articles = len(articleTexts)
idf = {
    word : math.log(num_articles / num_docs)
    for (word, num_docs)
    in num_docs_per_word.items()
}

print(idf['radiation'])
print(idf['blackbody'])
```

```python
tf_idf_matrix = []
for tf_vector in tf_matrix:
    tf_idf_vector = {
        word : tf * idf[word]
        for (word, tf)
        in tf_vector.items()
    }
    tf_idf_matrix.append(tf_idf_vector)
str(tf_idf_matrix)[:2000]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from numpy.lib.function_base import vectorize
# Create term frequency matrix
vectorizer = TfidfVectorizer(stop_words='english')
article_tfidf = vectorizer.fit_transform(articleTexts)

# Matrix of schools x number of words
print('shape', article_tfidf.shape)
print('vocab', vectorizer.get_feature_names_out()[1000:1200])
print(vectorizer.vocabulary_)

feature_names = vectorizer.get_feature_names_out()


article_tfidf = article_tfidf.toarray()

blackbody_index = vectorizer.vocabulary_.get(u'blackbody')



print(article_tfidf[black_body_radiation_index,blackbody_index])
print(article_tfidf[thermal_radiation_index,blackbody_index])
print(article_tfidf[civ_radiation_acc_index, blackbody_index])

```

```python
def sort_tfidf_entries_for_article(article_row):
    sorted_indices = np.argsort(article_row)[::-1]
    sorted_tfidf_values = np.take(article_row, sorted_indices)
    sorted_terms = [feature_names[i] for i in sorted_indices]
    return sorted_tfidf_values, sorted_terms



top_n = 5

for i in range(0, len(articleTexts)):
  sorted_values, sorted_terms = sort_tfidf_entries_for_article(article_tfidf[i])
  for j in range(0, top_n):
    print(f"{wikiRadiationList[i]}   {sorted_terms[j]}   {sorted_values[j]:.5f}")

# these also all make sense to me, compared with just tf, terms are even more specilized to thier articles, and they all seem to relate, i.e. fuel for nucular reactor, half,life,decay,expontetnial for half life etc..
```

```python
from math import sqrt
def cosine(X, Y):
    return np.sum(X * Y) / (sqrt(np.sum(X ** 2)) *sqrt(np.sum(Y ** 2)))

cosine_similarity = cosine(article_tfidf[black_body_radiation_index], article_tfidf[thermal_radiation_index]) # expected higher
print(cosine_similarity)

cosine_similarity = cosine(article_tfidf[alpha_particle_index], article_tfidf[thermal_radiation_index])# expected lower
print(cosine_similarity)

cosine_similarity = cosine(article_tfidf[alpha_particle_index], article_tfidf[beta_particle_index]) # expected higher
print(cosine_similarity)
```
